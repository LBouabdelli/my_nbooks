```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
data=pd.read_csv('avocado.csv')
```

```python
data.head(5)
```

```python
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
newdata=data.query('year==2015'and"region=='Albany'"and "type=='conventional'")[['Date','Total Bags','Small Bags']]
```

```python
newdata.head(5)
```

```python
data_melted=pd.melt(newdata,id_vars='Date',
                    value_vars=['Total Bags','Small Bags'],
                    var_name='Bags', value_name='Count')
```

```python
data_melted.head()
```

```python
sns.lineplot(data=data_melted, x='Date', y='Count', hue='Bags')
```

```python
data['Sales Volume'] = pd.qcut(data['Total Volume'], q=4,
    labels=['Poor','Modest','Good','Excellent'])
```

```python
data.head(5)
```

```python
cnt=data_melted.Count
```

```python
sns.countplot(data,x='year',hue='Sales Volume')
```

```python
melted_bags=pd.melt(data,id_vars='year',
                    value_vars=['Small Bags','Large Bags','XLarge Bags'],
                    var_name='Bags',value_name='Count')
```

```python
melted_bags.head()
```

```python
sns.relplot(data=melted_bags,kind='line',x='year',y='Count',hue='Bags')
```
